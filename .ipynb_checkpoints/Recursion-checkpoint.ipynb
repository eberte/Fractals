# Randomized tree "fractal"

Our first example of recursion will be a tree function, but with a twist. To start, our parameters are tortoise, or the placeholder turtle, length, or length of "branches", depth, or our depth of recursion, and angle, or angle of our branches. This tree's "branch" angles have been randomized so they can have a measure between 10 and 60 degrees using the random.randint command to make the angle measures between 10 and 60 degrees. Along with the angles, I have also randomized the length of the "branches" of the tree using the random.uniform command and limited the lengths from anywhere between 10 and 100 length units. For the base case, I said if my recursion depth is less than or equal to 1, I told the turtle to go forwards and backwards according to  the random length. For my recursive case, I told the turtle to set its speed to 0, or fastest, and to draw a forward line, and then turn left at a random angle. Each time the turtle gets the command turtle.lt or turtle.rt, I told the turtle to set the heading. I told the turtle to execute the tree function with a 2/3 length and a depth of d - 1 each time the turtle turned except for the final turtle.lt turn. 

```python
def tree(tortoise, length, depth, angle):
    length = random.uniform(10, 100)
    angle = random.randint(10, 60)  
    if depth <= 1:                     
        tortoise.forward(length)
        tortoise.backward(length)
    else:                              
        tortoise.speed(0)  
        tortoise.forward(length)
        tortoise.lt(angle)
        tortoise.setheading(angle) 
        tree(tortoise, length * (2 / 3), depth - 1, angle) 
        tortoise.right(angle)
        tortoise.setheading(angle)
        tree(tortoise, length * (2 / 3), depth - 1, angle) 
        tortoise.left(angle)
        tortoise.setheading(angle) 
        tortoise.backward(length)

        
```

The Main function incorporates the previously- written function and makes the previous function run. To start. I constructed my Turtle, so the program can show what it is doing, and I made the turtle turn left at a 90 degree angle.
I then defined my angle variable again, ran the function, and made the turtle get the screen commands for exiting on click, and then told screen to exit on click once the turtle was done, hence why it is at the end of the program.

```python
def main():     
    george = turtle.Turtle() 
    george.left(90)
    angle = random.randint(10, 60) 
    tree(george, 100, 9, angle)
    screen = george.getscreen() 
    screen.exitonclick() 
    
main()  
```

Here is an example of a randomized tree with a depth of 9.

# Quadratic Koch curve

Our second example of recursion is actually aesthetically pleasing this time. What is being made is a quadratic Koch curve, or a Koch curve with 90 degree turns instead of 60 degrees. The three parameters are tortoise, our placeholder Turtle, length, or the length of the curve, and depth, the depth of desired recursion. For my base case, I told the turtle to go forward if the depth of recursion was 0 by a specified length in the main loop. For my recursive case, I once again set the turtle speed to fastest, and ran the function with a length divided by 3. Depth - 1 also returns again. We run the function each time the turtle makes a lefr or right turn. 

```python
def quadkoch(tortoise, length, depth): 

    if depth == 0:                  
        tortoise.fd(length) 
    else:                       
        tortoise.speed(0) 
        quadkoch(tortoise, length / 3, depth - 1) 
        tortoise.lt(90)
        quadkoch(tortoise, length / 3, depth - 1) 
        tortoise.rt(90)
        quadkoch(tortoise, length / 3, depth - 1) 
        tortoise.rt(90)
        quadkoch(tortoise, length / 3, depth - 1) 
        tortoise.lt(90)
        quadkoch(tortoise, length / 3, depth - 1) 
```

I once again define a main function, construct the turtle, run the function, get screen, and specify exit on click. 

```python
def main(): 
    george = turtle.Turtle() 
    quadkoch(george, 400, 3) 
    screen = george.getscreen() 
    screen.exitonclick() 
main()
```

Here is an example of a quadratic Koch curve with a length of 400 and a depth of 3.

# 2 birds with one stone: Koch curve and Koch curve snowflake functions.

A Koch curve is a basic example of recursion. For this Koch curve function. its parameters are tortoise, our turtle placeholder, length, or the length of the curve, and depth, or depth of recursion. For the base case, I said if depth was 0, the turtle goes forward by a length specified in the Main loop. For the recursive case, I once again set the speed to fastest, ran the function exactly like the quadratic Koch function. I ran the function after a 60 degree left turn, a 120 degree right turn, and another 60 degree left turn. 

```python
def koch(tortoise, length, depth): 

    if depth == 0:              
        tortoise.fd(length)
    else:                   
        tortoise.speed(0) 
        koch(tortoise, length / 3, depth - 1)
        tortoise.lt(60)
        koch(tortoise, length / 3, depth - 1) 
        tortoise.rt(120)
        koch(tortoise, length / 3, depth - 1) 
        tortoise.lt(60)
        koch(tortoise, length / 3, depth - 1) 
```

I defined a main loop, constructed a turtle, ran the function, configured exit on click, and ran the main function. 

```python
def main():
    george = turtle.Turtle() 
    koch(george, 400, 3) 
    screen = george.getscreen() 
    screen.exitonclick() 
main()
```

A Koch curve with a length of 400 and a depth of 3. 

```python
def kochSnowFlake(tortoise, length, depth, sides): 


    for side in range(sides): 
        koch(tortoise, length, depth)
        angle = 360 / sides
        tortoise.rt(angle)
        turtle.setheading(angle)
```

```python
def main(): 
    sides = input('How many sides do you want the snowflake to have?') 
    bob = turtle.Turtle() 
    kochSnowFlake(bob, 400, 3, int (sides))
```

# Hilbert space- filling curve function

```python
def hilbert(tortoise, reverse, depth):
        
    if depth < 0:
        return

    if reverse == True:
        tortoise.rt(90)
        hilbert(tortoise, False, depth - 1)
        tortoise.fd(10)
        tortoise.lt(90)
        hilbert(tortoise, reverse, depth - 1)
        tortoise.fd(10)
        hilbert(tortoise, reverse, depth - 1)
        tortoise.lt(90)
        tortoise.fd(10)
        hilbert(tortoise, False, depth - 1)
        tortoise.rt(90)
        
    else:
        tortoise.lt(90)
        hilbert(tortoise, True, depth - 1)
        tortoise.fd(10)
        tortoise.rt(90)
        hilbert(tortoise, reverse, depth - 1)
        tortoise.fd(10)
        hilbert(tortoise, reverse, depth - 1)
        tortoise.rt(90)
        tortoise.fd(10)
        hilbert(tortoise, True, depth - 1)
        tortoise.lt(90)
        
```

```python
def main():
    reverse = True
    myTurtle = turtle.Turtle()
    myTurtle.lt(90)
    hilbert(myTurtle, reverse, 5)
    screen = myTurtle.getscreen()
    screen.exitonclick()
```
